# K-medoids 군집화 알고리즘
- 각 군집에서 대표 객체(medoids)를 임의로 찾음으로써 n개의 객체 중에서 k개의 군집을 찾는 것이다.
- 남은 각각의 객체는 가장 비슷한 medoids에 군집된다
- Step 1. n개의 객체 중 대표 객체(medoids)를 k개 지정한다.(n>k)
- Step 2. k개의 medoids를 지정 후, 나머지 객체를 유사성이 가장 높은 medoid에 할당한다. 여기서 유사성은 유클리디안 거리측도를 활용한다.
- Step 3. modoid가 아닌 다른 객체를 임의로 지정한다. 
- Step 4. 본래의 medoid와 임의의 medoid간 총 비용(cost)를 계산한다. 비용은 거리측도의 값을 모두 더한 것을 사용한다.
- Step 5. 비용이 낮은 쪽의 medoid와 교체한다.
- Step 6. 변화가 없을 때까지 Step 2 ~ Step 5를 반복한다.
- k-means에 비해 이상치에 덜 민감하지만 군집의 수와 데이터의 총 수에 비례하여 고비용으로 처리되게 된다는 단점이 있다.
- 본 코드는 k를 3개라고 가정한다.
- 유클리디안 거리측도를 활용한 cost에 추가적으로 Silhouette 평가 값도 구해본다.

```python
#거리를 구하는 함수 생성

def distance(df,centroid):
    for i in range(3): #만약 i=0이면 첫 번째 select_medoids의 대응되는 attribute와 Data들과의 거리를 구할 수 있다.
        #거리는 유클리디안 거리측도를 활용한다.
        df['distance_from_select_medoids{}'.format(i)]=( 
            np.sqrt(
            (df.iloc[:,0]-select_medoids[i][0])**2
            +(df.iloc[:,1]-select_medoids[i][1])**2
            +(df.iloc[:,2]-select_medoids[i][2])**2
            +(df.iloc[:,3]-select_medoids[i][3])**2
            )
         )
    return df
```

```python
df['cluster'] = 0 #어느 Cluster에 할당되었는지를 보여주기 위한 열 추가

#데이터들을 할당하는 함수 생성

def assignment(df):
    #세가지 값의 거리를 하나의 리스트에 대입
    distance_from_all = ['distance_from_select_medoids{}'.format(i) for i in range(3)]
    #그 값중 가장작은 값을 대입
    df['cluster'] = df.loc[:,distance_from_all].idxmin(axis=1) 
    #숫자부분만 남기기위해 lstrip으로 제거
    df['cluster'] = df['cluster'].map(lambda x: int(x.lstrip('distnace_fromselect_medoids'))) 
    return df
```

```python
#평가값을 구하는 함수 생성 (평가값은 거리측도의 합으로 결정한다.)

def Value_cal(df):
    distance_min=[]
    for i in range(150):
        distance_min.append(min(list(df.iloc[i,4:7]))) #거리중 가장 작은 값을 리스트에 추가
    a=sum(distance_min) #리스트들을 모두 더한다
    return a
```

## Silhouette
- 실루엣 값은 클러스터의 성능을 측정할 때 사용하는 지표이다.
- 한 클러스터 안의 데이터들이 다른 클러스터와 비교해서 얼마나 비슷한가를 나타낸다.
- 클러스터 안의 거리가 짧을 수록 좋고(cohesion), 다른 클러스터와의 거리는 멀수록 좋다(separation)
- 실루엣은 -1부터 1사이의 값을 가진다. 값이 높을 수록 성능이 좋다고 할 수 있다
- a(i) : 같은 클러스터 안에 있는 다른 데이터들과의 평균거리. a(i)는 i번째 데이터가 클러스터에 얼마나 잘 맞는지를 측정한다.(낮을 수록 현재 클러스터에 더 잘 속해있는 것)
- b(i) : i가 속하지 않은 다른 클러스터와의 평균 거리 중 가장 작은 거리이다. (클 수록 현재 클러스터에 더 잘 속해있는 것)

```python
#silhouette 평가 값 산출과정

for i in range(150): #i번재 데이터를 기준으로
    for j in range(150): #모든 데이터들과 거리를 구한다.
            z=np.sqrt(
            (df.iloc[i,0]-df.iloc[j,0])**2
            +(df.iloc[i,1]-df.iloc[j,1])**2
            +(df.iloc[i,2]-df.iloc[j,2])**2
            +(df.iloc[i,3]-df.iloc[j,3])**2)
            df.loc[j,'distance_from_data{}'.format(i)]=z
```

```python
#각 데이터들마다 자신이 속한 Cluster의 거리를 x리스트에 나타냄

x=[]
for i in range(150): 
    a=df.loc[i,'cluster']
    dis=[]
    for j in range(150): #모든 데이터들에 대해
            if df.loc[j,'cluster']==a: #같은 cluster인 경우.
                dis.append(df.iloc[j,8+i]) #dis리스트에 거리 값 추가(8번째 열이 distance_from_data0)
    x.append(sum(dis)) #거리 값들을 더한 값을 x리스트에 추가
```

```python
#각 데이터들마다 자신이 속하지 않은 cluster의 거리를 y리스트에 나타냄

y=[]
for i in range(150): #i는 몇 번째 데이터인지
    other_cluster={0:0,1:0,2:0} #복수개의 cluster와의 거리를 나타내기위해 딕셔너리자료형{key:Value} 선언
    a=df.loc[i,'cluster']
    dis=[]
    for j in range(150):
            if df.loc[j,'cluster']!=a: #속하지않은 cluster의 경우
                #딕셔너리[key]=Value 형태를 사용. 반복적으로 그 값을 더해나간다.
                other_cluster[df.loc[j,'cluster']]=other_cluster[df.loc[j,'cluster']]+df.iloc[j,8+i]
    #3개의 cluster를 딕셔너리로 선언했기에 만약 자신의 클러스터일경우를 대비 (min값에 영향)
    if other_cluster[0]== 0:
        other_cluster[0]=999999
    elif other_cluster[1]==0:
        other_cluster[1]=999999
    elif other_cluster[2]==0:
        other_cluster[2]=999999        
    y.append(min(other_cluster[0],other_cluster[1],other_cluster[2]))                    
```
